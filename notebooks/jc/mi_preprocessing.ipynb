```python
import numpy as np 
import pandas as pd
df = pd.read_csv('../../data/michigan_2019.csv')
counties = pd.read_csv('../../data/mi_counties.csv', names = ['fips', 'name'])
fips_dict = counties.set_index('name')['fips'].to_dict()
df['fips'] = df['county'].map(fips_dict)
df = df.dropna()
```

```python
county_list = df.fips.unique().tolist()
```

```python
def vax_percentage(df):
    percentage_list = []
    for fip in county_list:
        c_df = df[df.fips == fip]
        c_df['vaccinated_n'] = c_df['total_students']* c_df['vaccinated_perc']
        enrollment = c_df['total_students'].sum()
        percentage_vaccinated = (c_df['vaccinated_n'].sum()) / enrollment
        percentage_list.append(percentage_vaccinated)
    return percentage_list
```

```python
percentage_vaccinated = vax_percentage(df)
```

```python
mi_list = pd.DataFrame({'county': county_list, 'percentage_vaccinated': percentage_vaccinated})
```
