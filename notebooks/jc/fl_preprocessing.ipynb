```python
import numpy as np 
import pandas as pd
df = pd.read_csv('../../data/florida_2015.csv')
df2 = pd.read_csv('../../data/florida_2015.csv')
counties = pd.read_csv('../../data/fl_counties.csv', names = ['fips', 'name'])
fips_dict = counties.set_index('name')['fips'].to_dict()
df['county'] = df['county'].map(lambda x: x.rstrip('3'))
df['fips'] = df['county'].map(fips_dict)

df2['county'] = df2['county'].map(lambda x: x.rstrip('3'))
df2['fips'] = df2['county'].map(fips_dict)
```

```python
fl_vax_df_2015 = df[['fips', 'full_immunization_perc']]
fl_vax_df_2015['vax_percentage'] = fl_vax_df_2015['full_immunization_perc']
fl_vax_df_2015 = fl_vax_df_2015.drop('full_immunization_perc', axis=1)
fl_vax_df_2015['vax_percentage'] = fl_vax_df_2015['vax_percentage'].str.strip('%').astype(float)
fl_vax_df_2015['vax_percentage'] = fl_vax_df_2015['vax_percentage']/100
```

```python
fl_vax_df_2019 = df2[['fips', 'full_immunization_perc']]
fl_vax_df_2019['vax_percentage'] = fl_vax_df_2019['full_immunization_perc']
fl_vax_df_2019 = fl_vax_df_2019.drop('full_immunization_perc', axis=1)
fl_vax_df_2019['vax_percentage'] = fl_vax_df_2019['vax_percentage'].str.strip('%').astype(float)
fl_vax_df_2019['vax_percentage'] = fl_vax_df_2019['vax_percentage']/100
```
