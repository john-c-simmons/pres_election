```python
def vax_percentage(df):
    percentage_list = []
    for fip in df.fips.unique():
        c_df = df[df.fips == fip]
        #c_df['vaccinated_n'] = c_df['total_students']* c_df['vaccinated_perc']
        #enrollment = c_df['total_students'].sum()
        percentage_vaccinated = c_df['MMR (2 doses)'].sum()/len(c_df['MMR (2 doses)'])
        percentage_list.append(percentage_vaccinated)
    return percentage_list
```

```python
percentage_list2015 = vax_percentage(df)
fips_dict = counties.set_index('name')['fips'].to_dict()
tx_vax_df_2015 = pd.DataFrame(data=([df.fips.unique(), percentage_list2015])).T
tx_vax_df_2015 = tx_vax_df_2015.rename(columns={0:'fips', 1:'vax_percentage'})
tx_vax_df_2015.fips = tx_vax_df_2015.fips.astype(int)
```
