```python
df = pd.read_csv('quality_assurance.csv')
```

```python
df.head()
```

```python
# Target
y = df['flawed']
# Features
X = df.drop('flawed', axis=1)
```

#### Train-Test Split

Before performing any other preprocessing steps, split the data into training, validation, and testing sets.  Use the `train_test_split` utility function from sklearn with a `random_state` of 42 for reproducibility. **NOTE** You will be asked to perform cross-validation throughout this notebook. You will need *three sets of data* to do this successfully.

```python
# Initial Train-Test Split
X_tr, X_test, y_tr, y_test = train_test_split(X, y, test_size=.15, random_state=42)
```

```python
X_train, X_val, y_train, y_val = train_test_split(X_tr, y_tr, test_size=.15, random_state=42)
```

#### Missing Values

At least one column in this dataset contains missing values.  Use the sklearn `SimpleImputer` ([documentation here](https://scikit-learn.org/stable/modules/generated/sklearn.impute.SimpleImputer.html)) to fill in the missing values.  **Do not** add a "missing indicator".

If you are getting stuck at this step, you can drop the rows containing missing values, but this will mean your model performance will be worse. 

```python
X_train.isna().sum()
```

```python
X_val.isna().sum()
```

```python
# Instantiate a SimpleImputer
imputer = SimpleImputer(strategy='median')
```

```python
# Fit on training data
imputer.fit(X_train)
```

```python
# Transform training and validation data
X_train_imputed = pd.DataFrame(imputer.transform(X_train), columns=X.columns)
X_val_imputed = pd.DataFrame(imputer.transform(X_val), columns=X.columns)
X_test_imputed = pd.DataFrame(imputer.transform(X_test), columns=X.columns)
```

```python
print("X_train missing: ", X_train_imputed.isna().sum().sum())
print('X_val missing: ', X_val_imputed.isna().sum().sum())
print('X_test missing: ', X_test_imputed.isna().sum().sum())
```

#### Feature Scaling

Because we intend to use a model with regularization, the feature magnitudes need to be scaled in order to avoid overly penalizing the features that happen to have larger magnitudes.  Use the sklearn `StandardScaler` ([documentation here](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html)) to scale all of the features.  It is okay if you lose the feature names at this stage.

```python
cols = list(X.columns)
```

```python
# Instantiate StandardScaler
scaler = StandardScaler()
```

```python
# fit on Imputed training data
scaler.fit(X_train_imputed)
```

```python
# transform training and validation data
X_train_scaled = pd.DataFrame(scaler.transform(X_train_imputed), columns=cols)
X_val_scaled = pd.DataFrame(scaler.transform(X_val_imputed), columns=cols)
X_test_scaled = pd.DataFrame(scaler.transform(X_test_imputed), columns=cols)
```

```python
#MY step to rename scaled data
X_train_df = X_train_scaled
X_val_df = X_val_scaled
X_test_df = X_test_scaled
```

### Modeling

```python
def modeling_function(model, X_train, y_train, X_val, y_val):

    # fit model on training data
    model.fit(X_train, y_train)

    # make predictions on training and validation data
    train_preds = model.predict(X_train)
    val_preds = model.predict(X_val)

    # Print accuracy score
    print('Training recall: ', recall_score(y_train, train_preds))
    print('Validation recall: ', recall_score(y_val, val_preds))

    # return fitted model
    return model
```

```python
def model_test_evaluation(model, X_test, y_test):

    # make predictions on test data
    test_preds = model.predict(X_test)

    # Print accuracy score
    print('Test recall: ', recall_score(y_test, test_preds))
```
